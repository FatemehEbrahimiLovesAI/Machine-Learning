# Import Libraries

```python
import numpy as np
from math import sqrt,cos,radians
```

# Vectors

```python
v1 = np.array([3,4])
v2 = np.array([2,1])
```

# Vectors' Norms

```python
v1_n = sqrt( pow(v1[0],2) + pow(v1[1],2) )
v2_n = sqrt( pow(v2[0],2) + pow(v2[1],2) )

print(f"First Vector's Norm : {v1_n}\nSecound Vector's Norm : {v2_n}")
```

# Simple Dot Product 

```python
S_DP = v1[0]*v2[0]  + v1[1]*v2[1]

print(S_DP) 
```

# Beta ( Angle between two Vectors )

```python
beta = S_DP / (v1_n * v2_n)
print(beta)
```

# Cos Dot Product

```python
C_DP = v1_n * v2_n * cos(radians(beta)) # I noticed that in cos in Python math you must put the degree in radiants
print(C_DP)
```

# the Projection of first Vector on secound Vector

```python
v3_n = S_DP / v2_n
print(v3_n)
```

# Vector Direction

## Vector 1

```python
v1_VD = ( v1[0] / v1_n , v1[1] / v1_n )
v1_VD
```

## Vecor 2

```python
v2_VD = ( v2[0] / v2_n , v2[1] / v2_n )
v2_VD
```
