### Car Evolation with KNN classification
This project implements a K-Nearest Neighbors (KNN) classification model to predict the acceptability of cars based on various features.

#### Import Modules

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.neighbors import KNeighborsClassifier
```

#### Reading dataset file

```python
df = pd.read_csv(r'C:\Users\pc\Documents\university file\Machine Learning\Machine learning files\part5\Car Evolation\data.csv')
df.head()
```

#### Test and train split

```python
train,test = train_test_split(df, test_size=0.2,random_state=12)
print(f'train : \n{train.head()}\ntest : \n{test.head()}')
```

#### One-hot encoding

```python
one_hot_features = ['buying','maint']
train_one = train[one_hot_features]
test_one = test[one_hot_features]

one = OneHotEncoder(sparse_output=False)
train_one = one.fit_transform(train_one)
test_one = one.transform(test_one)
```

#### Ordinal encoding

```python
ordinal_features = ['doors','persons','lug_boot','safety']
train_ordinal = train[ordinal_features]
test_ordinal = test[ordinal_features]

categories = [
    ['2','3','4','5more'],
    ['2','4','more'],
    ['small','med','big'],
    ['low','med','high'],
]

ordinal = OrdinalEncoder(categories=categories)
train_ordinal = ordinal.fit_transform(train_ordinal)
test_ordinal = ordinal.transform(test_ordinal)
```

#### Reunification encoing and ordinal datas

```python
train_data = np.hstack((train_one,train_ordinal))
test_data = np.hstack((test_one,test_ordinal))
```

#### Seprating labels from data

```python
train_label = train.iloc[:,-1]
test_label = test.iloc[:,-1]
```

#### Create and fit model

```python
model = KNeighborsClassifier(7)
model.fit(train_data,train_label)
```

#### Prediction and accuracy 

```python
prediction = model.predict(test_data)
accuracy = accuracy_score(prediction, test_label)
accuracy
```
