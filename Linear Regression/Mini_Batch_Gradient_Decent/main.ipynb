```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
```

```python
dataset = fetch_california_housing()
```

```python
df = pd.DataFrame(data=dataset['data'],columns=dataset['feature_names'])
df.head()
```

```python
features = ['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup','Latitude','Longitude']

for feature in features:
    df[feature] = df[feature].replace(0,None)
    mean_value = np.mean(df[feature])
    df[feature] = df[feature].fillna(mean_value)

df.head()
```

```python
for feature in features:
    min_value = np.min(df[feature])
    max_value = np.max(df[feature])
    df[feature] = (df[feature] - min_value ) / (max_value - min_value)

df.head()
```

```python
X = df.iloc[:,:]
y = np.array(dataset['target'])
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,train_size=0.8,random_state=12)
```

```python
def gradient(X,y,y_hat):
    grads = []
    for xi in X.T:
        grads.append(2 * np.mean(xi * (y_hat - y)))
    return np.array(grads) 
```

```python
def update(w,eta,grads):
    return w - (eta * grads)
```

```python
def multi_variable_regression(X,w):
    return X @ w 
```

```python
def make_batches(X,y,batch_size):
    X_batches = []
    y_batches = []
    for i in range(0,len(X),batch_size):
        X_batches.append(X[i:i+batch_size])
        y_batches.append(y[i:i+batch_size])
    return np.array(X_batches),np.array(y_batches)
```

```python
def loss_batch(y_batch,y_hat):
    return np.mean((y_batch - y_hat) **2)
```

```python
m,n = X.shape
m,n
```

```python
batch_size = 8
X_batches, y_batches = make_batches(X_train,y_train,batch_size)
X_batches.shape , y_batches.shape , X_train.shape, y_train.shape
```

```python
epoches = 1000
eta = 0.1
w = np.random.randn(n)

for epoch in range(epoches):
    losses = []

    for (X_batch,y_batch) in zip(X_batches,y_batches):
        y_hat = multi_variable_regression(X_batch,w)
        loss = loss_batch(y_batch,y_hat)
        losses.append(loss)
        grads = gradient(X_batch,y_batch,y_hat)
        w = update(w,eta,grads)
    print(f'epoch : {epoch+1} mse: {np.mean(losses)}')
```

```python
def mse_score(y_batch,y_hat):
    return np.mean((y_batch - y_hat) **2)
```

```python
def mae_score(y_batch,y_hat):
    return np.mean(abs(y_batch - y_hat))
```

```python
def r2_score(y_batch, y_hat):
    return 1 - sum((y_batch - y_hat) ** 2) / sum((y_batch - y_batch.mean()) ** 2)
```

```python
batch_size = 32
X_batches, y_batches = make_batches(X_train,y_train,batch_size)
X_batches.shape , y_batches.shape , X_train.shape, y_train.shape
# w = np.random.randn(n)
w = [-1.8145861 ,  1.00559584, -0.34816822, -1.57737987, -1.42116979,
       -0.32263432, -1.10406993, -0.87049433]

epoches = 1000
eta = 0.1

main_mse_loss = []
main_mae_loss = []
main_r2_loss = []

for epoch in range(epoches):
    epoch_mse_loss = []
    epoch_mae_loss = []
    epoch_r2_loss = []

    for (X_batch,y_batch) in zip(X_batches,y_batches):

        y_hat = multi_variable_regression(X_batch,w)

        mse = mse_score(y_batch,y_hat)
        mae = mae_score(y_batch,y_hat)
        r2 = r2_score(y_batch,y_hat)

        epoch_mse_loss.append(mse)
        epoch_mae_loss.append(mae)
        epoch_r2_loss.append(r2)

        grads = gradient(X_batch,y_batch,y_hat)
        w = update(w,eta,grads)
    
    main_mse_loss.append(np.mean(epoch_mse_loss))
    main_mae_loss.append(np.mean(epoch_mae_loss))
    main_r2_loss.append(np.mean(epoch_r2_loss))

print(f"mse : {np.mean(main_mse_loss)}\nmae : {np.mean(main_mae_loss)}\nr2 : {np.mean(main_r2_loss)}")
```
